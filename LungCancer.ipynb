```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn import model_selection , metrics
from sklearn.svm import SVC
import seaborn as sn
import sklearn
```

```python

dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/survey lung cancer.csv')
dataset.info()
```

```python
dataset.describe(include='object')
```

```python
dataset['LUNG_CANCER'].hist()
```

```python
total = float(dataset.shape[0])
ploting = sn.countplot(x='GENDER', data=dataset)
for p in ploting.patches:
    height = p.get_height()
    ploting.text(p.get_x() + p.get_width()/2.,
           height,
           '{:.2f}'.format((height/total)*100),
           ha='center')
plt.legend()
plt.show()
```

```python
dataset['GENDER'] = dataset['GENDER'].map({'M':1,'F':0})
```

```python
dataset['LUNG_CANCER'] = dataset['LUNG_CANCER'].map({'YES':1,'NO':0})
```

```python


corr = dataset.corr()
features = corr.index
plt.figure(figsize=(10,10))
ax = sn.heatmap(dataset[features].corr(), annot=True)

```

```python
X = dataset.iloc[:,:-1]
y = dataset['LUNG_CANCER']
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=123)
```

```python


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train),columns=X_train.columns)
X_train_scaled

```

```python
X_test_scaled = pd.DataFrame(scaler.fit_transform(X_test),columns=X_test.columns)
X_test_scaled
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
accuracy = metrics.accuracy_score(y_test,prediction)
accuracy
```
